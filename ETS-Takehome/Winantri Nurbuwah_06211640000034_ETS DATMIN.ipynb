```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
data=pd.read_csv("online_retail.csv")
data.head()
```

#### DATA CLEANING

```python
np.sum(data.isnull())
```

```python
for i, j in data[data['Description'].isnull()].iterrows(): data['Description'][i]="Code-"+str(j['StockCode'])
```

```python
np.sum(data.isnull())
```

```python
z=data.groupby('InvoiceNo')
beli=[]
for i,j in z:
    beli.append(list(j['Description'].map(str)))
```

```python
dataa=pd.DataFrame(beli)
dataa
```

#### ASSOCIATION RULE

```python
association_rules=apriori(beli,min_support=0.005,
                         min_confidence=0.2,
                         min_lift=3, min_length=2)
association_result=list(association_rules)
```

```python
association_result[0]
```

```python
for item in association_result :
    #first index of the inner list
    pair = item[0]
    items = [a for a in pair]
    print("rule: " + items [0] + " ->" + items[1])
    
    #second ondex of the inner list
    print("support: " + str(item[1]))
    
    #third index of the list located at 10th
    #of the third index of the inner list
    
    print("cofidence: " + str(item[2][0][2]))
    print("lift: " + str(item[2][0][3]))
    print("=======================================")
```
