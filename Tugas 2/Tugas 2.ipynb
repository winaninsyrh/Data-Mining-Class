## Imputation Missing Value

```python
np.sum(data.isnull())
```

```python
data.isnull().sum()
```

```python
data.isnull().sum().sum()
```

```python
#imputasi missing value data numerik
data['Age'] = data['Age'].fillna((data['Age'].mean())) 
data['Type of Lesion'] = data['Type of Lesion'].fillna((data['Type of Lesion'].median()))
data.isnull().any()
```

```python
# imputasi untuk data numerik yang lain
data['Age']=data['Age'].fillna(99)
data['Type of Lesion.2']=data['Type of Lesion.2'].fillna(111)
data.isnull().any()
```

```python
#imputasi data kategorik dengan modus
data['Temperature of Extremities'] = data.fillna(data['Temperature of Extremities'].value_counts().index[0])
data.isnull().any()
```

```python
#imputasi data kategorik dengan modus sesuai kolom masing2
data = data.apply(lambda x:x.fillna(x.value_counts().index[0]))
data.isnull().any()
```

## Encode Labels

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Encode untuk Sex, Steroid, Anaroxia, Spleen Pabable, Spiders tanpa membuat kolom baru
data['Surgery'] = LabelEncoder().fit_transform(data['Surgery'])
data['Hospital Number'] = LabelEncoder().fit_transform(data['Abdomen'])
data['Pulse'] = LabelEncoder().fit_transform(data['Pulse'])
data['Temperature of Extremities'] = LabelEncoder().fit_transform(data['Temperature of Extremities'])
data['Outcome'] = LabelEncoder().fit_transform(data['Outcome'])
data['Type of Lesion'] = LabelEncoder().fit_transform(data['Type of Lesion'])
data.head()
```

## Standardize

```python
from sklearn.preprocessing import StandardScaler
```

```python
data['Age'] = StandardScaler().fit(data[['Age']]).transform(data[['Age']])
data['Type of Lesion'] = StandardScaler().fit(data[['Type of Lesion']]).transform(data[['Type of Lesion']])
data['Packed Cell volume'] = StandardScaler().fit(data[['Packed Cell volume']]).transform(data[['Packed Cell volume']])
data['Pulse'] = StandardScaler().fit(data[['Pulse']]).transform(data[['Pulse']])
data['Respiratory Rate'] = StandardScaler().fit(data[['Respiratory Rate']]).transform(data[['Respiratory Rate']])
data['Type of Lesion.2'] = StandardScaler().fit(data[['Type of Lesion.2']]).transform(data[['Type of Lesion.2']])
```
